```python
# imputation par la moyenne pour les valeurs quantitatives
```

```python
def imputation (df_imp, quantitatif = True, strategy = 'mean', missing_values = np.nan) :
    '''
        Fonction d'imputation
        - strategy prend l'une des valeurs mean, median, constant
    '''
    
    df_ql = df_imp.select_dtypes('object')
    df_qt = df_imp[df_imp.columns[df_imp.dtypes != 'object']]
    
    if len(df_qt.   columns) > 0 : 
        imp = SimpleImputer(missing_values = missing_values, strategy = strategy)
        imp.fit(df_qt.values[:, :])
        v = imp.transform(df_qt.loc[:, :])
        df_qt[df_qt.columns] = v
        
    if len(df_ql.columns) > 0:
        imp = SimpleImputer(missing_values = missing_values, strategy = 'most_frequent')
        imp.fit(df_ql.values[:, :])
        v = imp.transform(df_ql.loc[:, :])
        df_ql[df_ql.columns] = v
    
    df_r = df_qt.copy()
    df_r[df_ql.columns] = df_ql.values
    return df_r
```

```python
3. Gestion des outiliers
```

```python
def detect_outliers (df_detect_out) :
    q1 = np.quantile(df_detect_out, 0.25)
    q3 = np.quantile(df_detect_out, 0.75)
    eiq = q3 - q1
    li = q1 - (eiq * 1.5)
    ls = q3 + (eiq * 1.5)
    i = list(df_detect_out.index[(df_detect_out<li) | (df_detect_out>ls)])
    val = list(df_detect_out[i])
    return i, val

def gest_outliers (df_out, delete = False, strategy = "mean") :
    '''
        Gestion des valeurs aberrantes
    '''
    
    for col in df_out.columns[df_out.dtypes != 'object'] :
        print(f"colonnes {col}")
        i, val = detect_outliers(df_out[col])
        print(f"indices = {i}")
    
        if len(val) > 0 :
            if delete == False :
                print(f"val = {val}")
                for ind in i :
                    print(f" inV {i}")
                    if strategy == "mean" :
                        moy = np.mean(df[col])
                        df_out[col][ind] = moy
                    elif strategy == 'median' :
                        moy = np.median(df[col])
                        df_out[col][ind] = moy
            else:
                for ind in i :
                    df_out = df_out.drop(ind, axis=0)
                    
    return df_out
        
```

```python
def display (df) :
    plt.figure(figsize= (10, 5))
    df.boxplot()
display(df)
df = gest_outliers(df, delete=False)
```

```python
4. Encodage
```

```python
def encodeDf (df) :
    newdf = df.copy()
    for col in df.select_dtypes('object')  :
        enc_data = LabelBinarizer().fit_transform(df[[col]])
        if len(df[col].unique()) > 2 :
            myLabel = [f"{col}_{mod}" for mod in df[col].unique()]
        else :
            myLabel = [col]
            
        newdf = newdf.drop(col, axis=1)
        newdf[myLabel] = enc_data
    return newdf
```

```python
df = encodeDf(df)
df
```

```python
def normalize (df) :
    
    df = df.select_dtypes(include='number').apply(zscore)
    return df
```
